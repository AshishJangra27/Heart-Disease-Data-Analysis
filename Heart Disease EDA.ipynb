```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
  
df = pd.read_csv('heart.csv')
df.head()
```

## Distrubution Plot for Numeric Columns

```python
sns.displot(df['Age'], kde = True, color = 'blue')
```

```python
sns.displot(df['RestingBP'], kde = True, color = 'blue')
```

```python
sns.displot(df['Cholesterol'], kde = True, color = 'blue')
```

```python
sns.displot(df['MaxHR'], kde = True, color = 'blue')
```

## Pie Charts for Categorical

```python
df.groupby('Sex').size().plot(kind = 'pie', autopct = '%.1f')
```

```python
df.groupby('ChestPainType').size().plot(kind = 'pie', autopct = '%.1f')
```

```python
df.groupby('RestingECG').size().plot(kind = 'pie', autopct = '%.1f')
```

```python
df.groupby('ST_Slope').size().plot(kind = 'pie', autopct = '%.1f')
```

```python
df.groupby('HeartDisease').size().plot(kind = 'pie', autopct = '%.1f')
```
